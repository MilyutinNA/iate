# Расчет погрешности измерений

Результат всякого измерения всегда содержит некоторую погрешность. Поэтому в задачу измерений входит не только нахождение самой величины, но также и оценка допущенной при измерении погрешности.
Напомним, что **абсолютной погрешностью** приближенного числа называется разность между этим числом и его точным значением, причем ни точное значение, ни абсолютная погрешность принципиально неизвестны и подлежат оценке по результатам измерений.
**Относительной погрешностью** приближенного числа называется отношение абсолютной погрешности приближенного числа к самому этому числу.

Пусть проведено n измерений величины $x$. Тогда за лучшую оценку истинного значения результата измерений принимается среднее арифметическое значение

$<x> = \frac{1}{n}{\sum_{i=1}^n{x_i}$

где $x_i$ - результат i-го измерения.

Для оценки случайной погрешности измерения существует несколько способов. Наиболее распространена оценка с помощью стандартной или средней квадратичной погрешности $\sigma$ (ее часто называют стандартной погрешностью или стандартом измерений).
Средней квадратичной погрешностью называется величина

$S = \sqrt{\frac{\sum_{i=1}^n\left({<x>-x_{i}}\right)^2}{n-1}}$;

где $n$ - число наблюдений.
Сказанное выше о погрешностях относится к погрешностям отдельного измерения. Однако важнее знать, насколько может уклоняться от истинного значения $x$ среднее арифметическое $<x>$.
Теория показывает, что средняя квадратичная погрешность среднего арифметического равна:

$S = \frac{S_n}{\sqrt {n}}$

Если число наблюдений очень велико, то подверженная случайным колебаниям величина $S_n$ стремится к постоянному значению $\sigma$:

$\sigma = \lim_{n \rightarrow ∞} S_n$

Именно этот предел и входит в качестве параметра $\sigma$ в распределение Гаусса. Квадрат этой величины называется дисперсией измерений. В действительности, по результатам измерений всегда вычисляется не $\sigma$ , а ее приближенное значение $S_n$, которое, вообще говоря, тем ближе к $\sigma$ , чем больше $n$.

$f(\triangle x) = \frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{(\triangle x)^2}{2\sigma ^ 2}}$

параметр $\sigma$ определяет ширину распределения. Несколько кривых Гаусса для разных значений параметра $\sigma$ показаны на рисунке:
![распределение Гаусса](http://nuclphys.sinp.msu.ru/enc/images/e021p01_01.gif)

На рисунке видно, что абсолютная погрешность измерений представляется как некая $\triangle x$ и истинное значение входит в область $[x_{изм}-\triangle x;x_{изм}+\triangle x]$.

В случае лабораторных измерений, число n конечное, поэтому неверно считать за истину идеализированную кривую.
Определим доверительный интервал. Чем большим будет установлен этот интервал, тем с большей вероятностью $x_{ист}$ попадает в этот интервал. С другой стороны, более широкий интервал дает меньшую информацию относительно величины $x_{ист}$. Если ограничиться учетом только случайных погрешностей, то при небольшом числе измерений n для уровня доверительной вероятности $\alpha$ полуширина доверительного интервала равна:

$\triangle x_{сл} = t_{\alpha,n}S$

где $t_{\alpha,n}$ - коэффициент Стьюдента.


![коэффициенты Стьюдента](https://konspekta.net/lektsiinetimg/baza3/777970976022.files/image031.jpg)

Доверительная вероятность - вероятность того, что истинное значение находится в нашем полученном интервале. Следовательно, чем шире у нас интервал $[x_{изм}-\triangle x;x_{изм}+\triangle x]$, тем выше вероятность того, что истинное значение находится в этом интервале. Это видно из таблицы выше
![Доверительная вероятность](https://sun9-east.userapi.com/sun9-59/s/v1/ig2/3EFHC4X6XPOpObJRt-agyK-3asuZzcbxB6GgWW10ARy-U6fythrJc-xpMelhWAM6TJD1vCwG9t-DeWnpe-9JrEUq.jpg?size=709x471&quality=96&type=album)

Погрешности физических измерений принято подразделять на систематические, случайные и грубые.
* Систематические погрешности вызываются факторами, действующими одинаковым образом при многократном повторении одних и тех же измерений. Они скрыты в неточности самого инструмента и неучтенных факторах при разработке метода измерений. Систематическая погрешность, как правило, указана в паспорте или на шкале прибора, а в простейших случаях может быть принята равной половине цены деления младшего разряда шкалы.
* Случайные погрешности обязаны своим происхождением ряду причин, действие которых неодинаково в каждом опыте и не может быть учтено. Они имеют различные значения даже для измерений, выполненных одинаковым образом, то есть носят случайный характер.
* Третий тип погрешностей, с которыми приходится иметь дело - грубые погрешности или промахи. Под грубой погрешностью измерения понимается погрешность, существенно превышающая ожидаемую при данных условиях. Она может быть сделана вследствие неправильного применения прибора, неверной записи показаний прибора, ошибочно прочитанного отсчета, неучета множителя шкалы и т.п.

Будем считать что грубые погрешности исключены. Тем самым результирующая абсолютная погрешность будет иметь вид:

$\triangle x = \sqrt{\triangle x_{сист}^2 + \triangle x_{сл}^2}$

Очевидно, что при одном и том же значении $\triangle x$ результат может оказаться достаточно точным при измерении некоторой большой величины, тогда как при измерении малой величины его точность будет недостаточной.
Таким образом, становится понятной необходимость введения относительной погрешности, которая определяется как:

$\delta x =\frac{\triangle x}{<x>}$

и выражается, обычно, в процентах.

Рассмотрим, каким образом оценить случайную погрешность косвенно измеряемой величины $y$, которая является функцией некоторого числа m непосредственно измеряемых величин $x_{i}$, т.е. $y = f(x_{1},x_{2},...,x_{m})$
Само среднее значение $<y>$ можно найти из известной функциональной зависимости $y = f(x_{1},x_{2},...,x_{m})$, подставляя в качестве аргументов усредненные по всем проведенным опытам значения непосредственно измеренных величин $<x_{i}>$ . Соответствующие вычисления показывают, что абсолютная погрешность $\triangle y$ в этом случае определяется по формуле:
$$\triangle y = \sqrt{\sum_1^n\left(\frac{\partial f}{\partial x_{i}}\triangle x_{i}\right)^{2}}$$
где $\frac{\partial f}{\partial x_{i}}$ - частная производная
Частная производная - это такая производная, которая вычисляется от функции f по аргументу $<x_{i}>$, притом как все остальные аргументы считаются постоянными.

Относительная погрешность для косвенно измеряемой величины y определяется как:
$$\delta y = \sqrt{\sum_1^n\left(\frac{\partial ln f}{\partial x_{i}}\triangle x_{i}\right)^{2}}$$

Учитывая простую связь между абсолютной и относительной погрешностями $\delta y =\frac{\triangle y}{<y>}$, легко по известной величине $\triangle y$ вычислить $\delta y$ и наоборот

# Резюмируем

При прямых (непосредственных) измерениях
1. Вычисляется среднее из n измерений:

$<x> = \frac{1}{n}{\sum_{i=1}^n{x_i}$;

2. Определяется среднеквадратичная погрешность среднего арифметического:

$S = \sqrt{\frac{\sum_{i=1}^n\left({<x>-x_{i}}\right)^2}{n(n-1)}}$;

3. Задается доверительная вероятность $\alpha$ и определяется коэффициент Стьюдента $t_{\alpha,n}$ для заданного $\alpha$ и числа произведенных измерений $n$ по таблице;
4. Находится полуширина доверительного интервала (абсолютная погрешность результата измерений):

$\triangle x = \sqrt{\triangle x_{сист}^2 + \triangle x_{сл}^2}$, где $\triangle x_{сл} = t_{\alpha,n}S$

5. Оценивается относительная погрешность результата измерений:

$\delta x =\frac{\triangle x}{<x>}$;

6. Окончательный результат записывается в виде:

$x = <x> \pm \triangle y$.


При косвенных измерениях:

1. для каждой серии измерений величин, входящих в определение искомой величины, производится обработка в описанной выше последовательности (пункты 1-6 при прямых измерениях). При этом для всех измеряемых величин задают одно и то же значение доверительной вероятности $\alpha$;

2. определения искомой величины через средние значения $<x_{i}>$ как:

 $<y> = f(<x_{1}>,<x_{2}>,...,<x_{m}>)$;

3. Оценивается точность результата косвенных измерений по формуле:

$\triangle y = \sqrt{\sum_{i=1}^n\left(\frac{\partial f}{\partial x_{i}}\triangle x_{i}\right)^{2}}$, либо: $\delta y = \sqrt{\sum_{i=1}^n\left(\frac{\partial ln f}{\partial x_{i}}\triangle x_{i}\right)^{2}}$, где производные вычисляются при средних значениях величин;

4. Определяется относительная, либо абсолютная погрешность результата серии косвенных измерений через формулу их связи:

$\delta y =\frac{\triangle y}{<y>}$;

5. Окончательный результат записывается в виде:

$y = <y> \pm \triangle y$, т.е. $y = f(x_{1},x_{2},...,x_{m})$.

Возможен и другой подход (оба способа при этом дают близкие результаты):
1. для каждого выполненного опыта вычислить:

$y_{i} = f(x_{1}^{(i)},x_{2}^{(i)},...,x_{m}^{(i)})$;

2. найти $<y>$ как среднее арифметическое:

$<y> = \frac{1}{n}{\sum_{i=1}^n{y_i}$;

3. найти абсолютную погрешность $\triangle y$ по формуле:

$ \triangle y_{сл} = t_{\alpha,n}S$,

где $t_{\alpha,n}$ - коэффициент Стьюдента (табличное значение),
$S$ - средняя квадратичная погрешность среднего арифметического:
$S = \sqrt{\frac{\sum_{i=1}^n\left({<y>-y_{i}}\right)^2}{n(n-1)}}$;


4. найти относительную погрешность

$\delta y = \sqrt{\sum_{i=1}^n\left(\frac{\partial ln f}{\partial x_{i}}\triangle x_{i}\right)^{2}}$.

#### Решение дифференциала с Sympy Diff ()
Для дифференциации Sympy предоставляет нам различать способ вывода производной функции.

Предположим, у нас есть функция: F(x) = х²
Производная функции от x: f‘(x) = 2x.
Посмотрим, как мы можем достичь этого, используя Sympy.

```python
import sympy as sym

x = sym.Symbol('x')
f = x**2

derivative_f = f.diff(x)
derivative_f
```

#### Решение производных в Python
Теперь для расчета производной функции в конкретной точке на Sympy есть lambdify функция.
Предположим, у нас есть функция: F(x) = х²
Производная функции от x: f‘(x) = 2x
Значение производная функции от x при х=3: F(3) = 9, f‘(3) = 6

```python
import sympy as sym

x = sym.Symbol('x')
f = x**2

derivative_f = f.diff(x)
f1 = sym.lambdify(x, derivative_f)
f1(3)
```

```python
f = x**2
f2 = sym.lambdify(x, f)
f2(3)
```

#### Основные производные правила в Python Sympy

Есть определенные правила, которые мы можем использовать для расчета производной дифференцируемых функций.

1. Правило питания:
$$f′(x^{n})^{(N-1)}$$

```python
import sympy as sym

x = sym.Symbol('x')
f = x**5

derivative_f = f.diff(x)
derivative_f
```

2. Правило продукта:
$$(uv)′ = u′v + uv′$$

```python
import sympy as sym

x = sym.Symbol('x')
f = sym.exp(x)*sym.cos(x)
derivative_f = f.diff(x)
derivative_f
```

Обратите внимание - в случае выше мы импортировали всю библиотеку sympy и задали ей короткое название sym. Вызов функций из библиотеки производится только в таком формате - lib.func_name_from_sym(argument).
Но на самом деле можно обойтись укороченной записью func_name_from_sym(argument) без указания библиотеки перед функцией, однако будьте внимательны при создании своих функций. Если ваша функция по названию будет совпадать с функцией из импортированной библиотеки, то будет браться последняя.
Чтобы не указывать имя библиотеки перед функцией - необходимо задать импорт в следующем виде:

```python
from sympy import *

x = Symbol('x')
f = exp(x) * cos(x)
derivative_f = f.diff(x)
derivative_f
```

Как видно ниже - такой формат импорта библиотеки не позволяет нам определить функцию через lib.func_name_from_sym(argument), и при задании func_name_from_sym(argument) вызывается ближайшая перед вызовом функция.

```python
from sympy import *

def Symbol(str):
    print('Вызов нашей функции')

x = Symbol('x')
x = sympy.Symbol('x')
```

3. Правило цепочки:
$$(g(x)) \cdot g′(x)$$

```python
import sympy as sym
x = sym.Symbol('x')
f = sym.cos(x**2)
derivative_f = f.diff(x)
derivative_f
```

#### Производные многомерные функции с использованием Sympy
Примеры, которые мы видели выше, имели одну переменную. Но мы более склонны столкнуться с функциями, имеющими более одной переменной в них. Такие производные, как правило, называются частичными производными.


```python
import sympy as sym

x , y = sym.symbols('x y')
f = x**4+x*y**4

derivative_f = f.diff(x)
derivative_f
```

Мы можем частично дифференцировать функцию сначала по x, а затем по y.

```python
import sympy as sym

x , y = sym.symbols('x y')
f = x**4+x*y**4

derivative_f = f.diff(x,y)
derivative_f
```
