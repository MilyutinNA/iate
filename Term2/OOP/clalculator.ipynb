```python
from tkinter import *

root = Tk()
root.title('Calculator')

e = Entry(root, width=35, borderwidth=5)
e.grid(row=0,column=0,columnspan=3, padx=10, pady=10)

def button_click(number):
    current = e.get()
    e.delete(0,END)
    e.insert(0,str(current) + str(number))

def button_clear():
    e.delete(0,END)

def button_add():
    first_number = e.get()
    global f_num
    f_num = int(first_number)
    e.delete(0,END)

def button_equal():
    second_number = e.get()
    e.delete(0,END)
    e.insert(0, f_num + int(second_number))


button_1 = Button(root, text='1',padx=40, pady=20, command=lambda: button_click(1))
button_2 = Button(root, text='2',padx=40, pady=20, command=lambda: button_click(2))
button_3 = Button(root, text='3',padx=40, pady=20, command=lambda: button_click(3))
button_4 = Button(root, text='4',padx=40, pady=20, command=lambda: button_click(4))
button_5 = Button(root, text='5',padx=40, pady=20, command=lambda: button_click(5))
button_6 = Button(root, text='6',padx=40, pady=20, command=lambda: button_click(6))
button_7 = Button(root, text='7',padx=40, pady=20, command=lambda: button_click(7))
button_8 = Button(root, text='8',padx=40, pady=20, command=lambda: button_click(8))
button_9 = Button(root, text='9',padx=40, pady=20, command=lambda: button_click(9))
button_0 = Button(root, text='0',padx=40, pady=20, command=lambda: button_click(0))
button_add = Button(root, text='+',padx=39, pady=20, command=button_add)
button_equal = Button(root, text='=',padx=91, pady=20, command=button_equal)
button_clear = Button(root, text='Clear',padx=79, pady=20, command=button_clear)

button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)

button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)

button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)

button_0.grid(row=4, column=0)
button_clear.grid(row=4, column=1, columnspan=2)
button_add.grid(row=5, column=0)
button_equal.grid(row=5, column=1, columnspan=2)

root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk

class Calc(Tk):

    def __init__(self):
        super().__init__()
        self.title('Calculator')
        self.put_frames()
        self.put_menu()

    def put_frames(self):
        self.main_page = MainPage().grid(row=0,column=0)
        self.additional_func = TypesCalcs().grid(row=1, column=0)

    def put_menu(self):
        self.config(menu=MainMenu(self))

    def get_info_page(self):
        all_frames = [f for f in self.children]
        for f_name in all_frames:
            self.nametowidget(f_name).destroy()
        self.info_page = InfoPage().grid(row=0, column=0)


class CountRows(Frame):

    num_columns = 3
    num_row = -1

    def count_row(self):
        self.num_row = self.num_row + 1
        return self.num_row

class MainMenu(Menu):
    def __init__(self, mainwindow):
        super().__init__(mainwindow)

        file_menu = Menu(self)
        options_menu = Menu(self)

        file_menu.add_command(label="Refresh")
        file_menu.add_separator()
        file_menu.add_command(label="Quit")

        options_menu.add_command(label="Switch language")
        options_menu.add_separator()
        options_menu.add_command(label="Switch theme")

        self.add_cascade(label="File")
        self.add_cascade(label="Options")


class TypesCalcs(CountRows):

    def __init__(self):
        super().__init__()
        self.l = Label(self, text = 'Тип калькулятора')
        self.l.grid(row=self.count_row(),column=0,columnspan=1)
        self.com = ttk.Combobox(self, values=['Обычный', 'Инженерный','Построение графиков'])
        self.com.grid(row=self.count_row(),column=0,columnspan=1)


class InfoPage(Frame):

    def __init__(self):
        super().__init__()
        self.l = Label(self, text = 'Временно недоступно\nНа данный момент, калькулятор\nне поддерживает операцию сложения\nПопробуйте позже', font=30)
        self.l.grid(row=0,column=0,columnspan=1)
        self.b = Button(self,text = 'Вернуться назад', command=self.return_back)
        self.b.grid(row=1,column=0,columnspan=1)

    def return_back(self):
        all_frames = [f for f in self.children]
        for f_name in all_frames:
            self.nametowidget(f_name).destroy()
        self.master.put_frames()


class MainPage(CountRows):

    buttons_list = [[7, 8, 9], [4, 5, 6], [1, 2, 3]]

    def __init__(self):
        super().__init__()
        self.grid_calc_nums = {}
        self.e = Entry(self, width=35, borderwidth=5)
        self.e.grid(row=self.count_row(),column=0,columnspan=3, padx=10, pady=10)

        self.buttons_grid()

        self.button_0 = Button(self, text='0',padx=40, pady=20, command=lambda: self.button_click(0))
        self.button_add = Button(self, text='+',padx=39, pady=20, command=self.button_add)
        self.button_equal = Button(self, text='=',padx=91, pady=20, command=self.button_equal)
        self.button_clear = Button(self, text='Clear',padx=79, pady=20, command=self.button_clear)

        self.button_0.grid(row=4, column=0)
        self.button_clear.grid(row=4, column=1, columnspan=2)
        self.button_add.grid(row=5, column=0)
        self.button_equal.grid(row=5, column=1, columnspan=2)

    def buttons_grid(self):
        for r in self.buttons_list:
            tup_num_row = self.count_row()
            for i,c in enumerate(r):
                self.grid_calc_nums[c] = Button(self, text=f'{c}',padx=40, pady=20, command=lambda x = c: self.button_click(x))
                self.grid_calc_nums[c].grid(row=tup_num_row, column=i)

    def button_click(self, number):
        current = self.e.get()
        self.e.delete(0,END)
        self.e.insert(0,str(current) + str(number))

    def button_clear(self):
        self.e.delete(0,END)

    def button_add(self):
        self.first_number = self.e.get()
        self.f_num = int(self.first_number)
        self.e.delete(0,END)
        self.master.get_info_page()

    def button_equal(self):
        self.second_number = self.e.get()
        self.e.delete(0,END)
        self.e.insert(0, self.f_num + int(self.second_number))


c = Calc()
c.mainloop()
```
